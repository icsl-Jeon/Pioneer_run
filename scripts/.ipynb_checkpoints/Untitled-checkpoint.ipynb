```python
#!/usr/bin/env python
import rosbag
import tf
import tf.transformations as tft
import matplotlib.pyplot as plt
from argparse import ArgumentParser
import numpy as np
from geometry_msgs.msg import Point 


t_ref = []
v_ref = []
w_ref = [] 
bagfile = "2018-09-06-20-48-16.bag"


bag=rosbag.Bag(bagfile)
for topic,msg,t in bag.read_messages(topics=['/target/cmd_vel_mux/input/teleop']):
    v_ref.append(msg.linear.x)
    w_ref.append(msg.angular.z)
    t_ref.append(t.secs + t.nsecs*1e-9)
    
    
plt.subplot(211)
plt.plot(t_ref,v_ref,'r',label='x_ref')
plt.title('target v')
plt.legend()
plt.subplot(212)
plt.plot(t_ref,w_ref,'r',label='w_ref')
plt.title('target w')
plt.legend()
plt.show()
```
